```python
5:ans:-     BOX PLOT:-
                        A box plot, also known as a box-and-whisker plot, is a graphical 
                        representation of a dataset that shows the distribution of the data,
                        including the median, quartiles, and outliers.
            
            USES:-
                        Box plots are used to visually summarize and compare the distribution of a dataset,
                        especially when comparing multiple datasets. They provide information about the central
                        tendency, spread, skewness, and presence of outliers in the data.
                        
                (CODES ARE BELOW)
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
box1 = [np.random.normal(100,10,200) for std in range(1, 4)]
fig, ax = plt.subplots()
ax.boxplot(box1)
ax.set_xlabel('box1')
ax.set_ylabel('Value')
ax.set_title('Box plot1 example')
plt.show()
```

```python
box2 = [np.random.normal(90,20,200) for std in range(1, 4)]
fig, ax = plt.subplots()
ax.boxplot(box2)
ax.set_xlabel('box2')
ax.set_ylabel('Value')
ax.set_title('Box plot12 example')
plt.show()
```
